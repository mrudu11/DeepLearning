```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

#loading the MNIST dataset
(x_train,y_train),(x_test,y_test)=keras.datasets.mnist.load_data()
```

```python
#analyzing the image
print(x_train)

#for checking the size
print(x_train.shape)

print(y_train)

#for checking the size of the y_train
print(y_train.shape)

print(x_test)
#for checking the size of the x_test
print(x_test.shape)

print(y_test)
#for checking the size of
print(y_test.shape)
```

```python
#preprocesing the data
x_train =x_train.reshape(-1,784).astype("float32")/255.0
x_test = x_test.reshape(-1,784).astype("float32")/255.0
```

```python
#ONE HOT ENCODING
y_train = keras.utils.to_categorical(y_train,10)
y_test = keras.utils.to_categorical(y_test,10)
```

```python
#developing the model
ann = tf.keras.models.Sequential()

#adding 1st layer
ann.add(tf.keras.layers.Dense(units=3, activation='relu'))

#adding 2nd layer
ann.add(tf.keras.layers.Dense(units=3, activation='relu'))

#adding 3rd layer
ann.add(tf.keras.layers.Dense(units=3, activation='relu'))

#adding 4th layer
ann.add(tf.keras.layers.Dense(units=3, activation='relu'))

#adding 5th layer
ann.add(tf.keras.layers.Dense(units=3, activation='relu'))

#adding 6th layer
ann.add(tf.keras.layers.Dense(units=3, activation='relu'))

#adding 7th layer
ann.add(tf.keras.layers.Dense(units=3, activation='relu'))

#adding 8th layer
ann.add(tf.keras.layers.Dense(units=3, activation='relu'))

#adding 9th layer
ann.add(tf.keras.layers.Dense(units=3, activation='relu'))

#adding output layer
#here if our model has the output based in 10 columns that why we need to add the 10 neurons inside the output layer
#means if we have to choose the only one output then use softmax and if we had to choose the multiple outputs then use sigmoid
ann.add(tf.keras.layers.Dense(units=10 , activation='softmax'))
```

```python
#using the optimizer to run the code
ann.compile(optimizer='adam', loss='mean_squared_error',metrics=['accuracy'])

#to train the model
history = ann.fit(x_train, y_train, batch_size = 32, epochs = 10 , validation_data=(x_test, y_test))
```

```python
test_loss, test_accuracy = ann.evaluate(x_test, y_test, verbose=0)
print("Test Accuracy:", test_accuracy)
```

```python
#using the optimizer to run the code
#Categorical cross-entropy punishes the model if it gives a low probability to the correct class label, and rewards it when it gives a high probability to the correct class label, so the model can clearly distinguish the correct label among multiple classes.
ann.compile(optimizer='sgd', loss='categorical_crossentropy',metrics=['accuracy'])

#to train the model
ann.fit(x_train, y_train, batch_size = 32, epochs = 40)
```

```python

plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Test Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
