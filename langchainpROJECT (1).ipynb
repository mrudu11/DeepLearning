```python
#import
from transformers import pipeline
from langchain_community.llms import HuggingFacePipeline
from langchain_core.prompts import PromptTemplate
```

```python
#model
generator = pipeline(
    "text-generation",
    model="google/flan-t5-base",      #SUMMERIZATION AND TRANSLATION and text generation
    max_length=200
)
```

```python
from langchain_community.llms import HuggingFacePipeline

llm = HuggingFacePipeline(pipeline=generator)
```

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate.from_template(
"""
You are a professional email writer.

Write a {tone} email for the following request.

Request: {task}

Format:
Subject: <subject>
Dear Sir/Madam,
<email body>
Thank you.
Sincerely,
[Your Name]

Only return the email.
"""
)
```

```python
chain = prompt | llm
```

```python
task = input("Enter your request: ")
tone = input("Enter tone: ")

result = chain.invoke({"task": task, "tone": tone})

output = result.split("Subject:")[-1]   # take only email part
output = "Subject:" + output.strip()

print("\n----- GENERATED EMAIL -----\n")
print(output)
```
