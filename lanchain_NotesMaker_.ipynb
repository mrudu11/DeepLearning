```python
#text generation
#use AI models without manually loading tokenizer and model.
from transformers import pipeline

generator = pipeline(
    "text-generation",
    model="tiiuae/falcon-7b-instruct",   # better model
    max_new_tokens=100
)
```

```python
def generate_notes(topic):
    prompt = f"""
Explain {topic} in simple words.
Give ONLY 3 points.
Number them 1, 2, 3.
Stop after 3 points.

Answer:
"""

    result = generator(
        prompt,
        max_new_tokens=80,
        temperature=0.5,
        do_sample=False   # more controlled output
    )[0]['generated_text']

    clean_output = result.replace(prompt, "").strip()

    # Optional: cut extra text
    lines = clean_output.split("\n")
    return "\n".join(lines[:3])
```

```python
topic = input("Enter topic: ")

output = generate_notes(topic)

print("\n----- STUDY NOTES -----\n")
print(output)
```
